```python
import requests
from bs4 import BeautifulSoup
```

```python
web=requests.get("https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets")
web
```

```python
web.content
```

```python
soup=BeautifulSoup(web.text,"html.parser")
soup
```

```python
## Total Products
boxes=soup.find_all("div",class_="col-md-4 col-xl-4 col-lg-4")
len(boxes)
```

```python
### Getting Names
names=soup.find_all("a",class_="title")
names
```

```python
for i in names:
    print(i.text)
```

```python
## Getting Prices
prices=soup.find_all("h4",class_="price float-end card-title pull-right")
prices
```

```python
for i in prices:
    print(i.text)
```
